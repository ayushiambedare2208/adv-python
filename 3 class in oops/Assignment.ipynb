HOW DOES A TRIANGLE FORM
1. Traingle inequality sum of any two side should be greater than third side
2. Perimeter is sum of all sides of a trainagle 
3. Area is calculated by herons formula

```python
from dataclasses import dataclass
```

```python
@dataclass
class Triangle:
    a : int | float
    b : int | float
    c : int | float
    
    def is_triangle(self):
        if ((self.a + self.b > self.c) and
            (self.b + self.c > self.a) and
            (self.c + self.a > self.b)):
            return True
        else:
            return False
        
    def perimeter(self):
            if self.is_triangle():
                p = self.a + self.b + self.c
                return p
            else:
                raise ValueError(f"Given sides {self.a}, {self.b}, {self.c} do not form a triangle")
        
    def area(self):
            s = self.perimeter()/2
            a = (s*(s-self.a)*(s-self.b)*(s-self.c))**1/2
            return a

```

Calculating perimeter and area for multiple traingels

```python
T1=Triangle(3,4,5)
T2=Triangle(12,13,15)
T3=Triangle(4,5,7)
T4=Triangle(1,2,2)
T5=Triangle(1,2,4)
```

```python
t = [T1, T2, T3, T4, T5]
t
```

```python
for i in t:
    print(i)
    if i.is_triangle():
        print(f"Perimeter of the triangel is {i.perimeter()}")
        print(f"Area of triangle is {i.area():.2f}")
    else:
        print("Doesn't form a triangel")

    print("/n-------------------------/n")
```
