POLYMORPHISM IN USER DEFINED CLASSES ALSO CALLED AS METHOD OVERRIDING

```python
class India:

    def capital(self):
        print("New Delhi is the capital city of India")

    def language(self):
        print("Hindi is the most spoken language in India")
```

```python
class USA:
    
    def capital(self):
        print("Washingtn DC is the capital city of USA")

    def language(self):
        print("English is widely spoken language in USA")
```

```python
class France:
    
    def capital(self):
        print("Paris is the capital city of France")
    
    def language(self):
        print("French is widely spoken language in France")
```

```python
c1 = India()
type(c1)
```

```python
c1.capital()
```

```python
c1.language()
```

```python
c2 = France()
type(c2)
```

```python
c2.capital()
```

```python
c2.language()
```

 CALCULATE AREA AND PERIMETER FOR DIFFERENT SHAPES

```python
from dataclasses import dataclass
```

```python
@dataclass
class Triangle:
    a : int | float
    b : int | float
    c : int | float
    
    def is_triangle(self):
        if ((self.a + self.b > self.c) and
            (self.b + self.c > self.a) and
            (self.c + self.a > self.b)):
            return True
        else:
            return False
        
    def perimeter(self):
            if self.is_triangle():
                p = self.a + self.b + self.c
                return p
            else:
                raise ValueError(f"Given sides {self.a}, {self.b}, {self.c} do not form a triangle")
        
    def area(self):
            s = self.perimeter()/2
            a = (s*(s-self.a)*(s-self.b)*(s-self.c))**1/2
            return a
```

```python
@ dataclass
class Rectangle:
    height : int | float
    width : int | float

    def perimeter(self):
        return 2 * (self.height + self.width)
    
    def area(self):
        return self.height*self.width
```

```python
@dataclass
class Circle:
    radius: int | float

    def perimeter(self):
        return 2 * 3.14 * self.radius
    
    def area(self):
        return 3.14 * (self.radius**2)
```

```python
shapes = (
    Triangle( 12, 14, 17),
    Circle(12),
    Rectangle(14, 12)
)
```

```python
for i in shapes:
    print(i, type(i))
    print(f"Perimeter : {i.perimeter():.2f}")
    print(f"Area : {i.area():.2f}")
    print("\n" ,"-----------"*5, "\n")
```
