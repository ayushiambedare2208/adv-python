# Multiplying 3 to a list vs array

```python
d = [11, 12, 13, 14, 15]
type(d)
```

```python
d*3
#it will repeat the list 3 times but we want to multiply each element of list by 3 
```

```python
# In list we need to apply loop for calculations
e = []
for i in d:
    e.append(i*3)
```

```python
e
```

```python
# Multiplying using Numpy Arrays 
d1 = np.array(d)
d1
```

```python
type(d1)
```

```python
# BOARDCASTING IN NUMPY ARRAYS
e1 = d1*3
```

```python
e1
```

```python
d1 + 5
```

```python
4*d1 + 5
```

#  Products of element inside list vs numpy array

```python
g = [51, 101, 48, 49, 50, 53]
type(g)
```

```python
# for list need to apply for loop
p = 1
for i in g:
    p = p*i
```

```python
p
```

```python
sum(g)
```

```python
# numpy array use np.prod
p1 = np.prod(g)
p1
```
